```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(
    "/content/drive/MyDrive/Introduction to Machine Learning with Python ( PDFDrive.com )-min.pdf"
)

documents = loader.load()

print("PDF Loaded Successfully")
```

```python
import google.generativeai as genai

from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

# Gemini API Key
API_KEY = "AIzaSyD0a-arwYvAGXHu0YBM9FRsjzs48-rIcmM"

# Configure Gemini
genai.configure(api_key=API_KEY)

# Split PDF text
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

texts = text_splitter.split_documents(documents)

# Create embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
# Store vectors
vector_store = FAISS.from_documents(texts, embeddings)

# Ask question
query = input("Ask Question From PDF: ")

# Similarity search
docs = vector_store.similarity_search(query)

from transformers import pipeline

qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

question = input("Ask Question From PDF: ")

context = docs[0].page_content

result = qa_pipeline(
    question=question,
    context=context
)

print("\nAnswer:\n")
print(result["answer"])
```
